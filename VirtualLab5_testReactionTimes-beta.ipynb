If the class question data you'd like to look at is **continuous**, we should use a scatterplot. Here, we'll also see if visual reaction times correlate with auditory reaction times. The function below, `sns.regplot()` will plot a scatterplot with a linear regression line. The argument `ci=68` means that this function will plot the data with a 68% confidence interval. (Why 68%?)

````python
from statsmodels.formula.api import ols


fig,ax = plt.subplots(1,3,figsize=(9,3))

category = 'Sleep'

sns.regplot(x='Auditory',y='Visual',data=data,ax=ax[0],ci=68)
sns.regplot(x=category,y='Visual',data=data,ax=ax[1],ci=68)
sns.regplot(x=category,y='Visual',data=data,ax=ax[2],ci=68)

fig.tight_layout()
plt.show()
````

For this type of analysis, in which we're asking about the relationship between two continuous variables, we can compute a Pearson's correlation coefficient (*r*) to see if the data is significantly linearly correlated. The function below will compare two variables of the same length and compute a Pearson's *r* as well as a p-value.

````python
r, p = stats.pearsonr(data['Auditory'], data['Sleep'])
print('Pearsons r: ',r)
print('P-value: ',p)
````
